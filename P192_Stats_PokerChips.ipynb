# Fair and Biased Poker Chips
Normal poker chips can land with either a black face up or a white face up. Biased chips always land with a black face up.

More the experiment: [Lab Notebook](https://www.overleaf.com/read/yksrdzwsfqnq)

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('bmh')
%matplotlib inline
import math
import statistics
```

**Simulations**


```python
def shake_box(A_arr):                   
  """
    input: array containing no. of normal (black-and-white) chips; 
    returns: no. of chips that land black face up 
  """
  def black_count(A):                     # returns the no. of chips that land black face up

    x = np.zeros(A)
    black = 0       
    for i in range(A):
      x[i] =  np.random.randint(2)        # 2 possible outcomes: black = 0; white = 1
      if x[i] == 0:                
        black+=1

    black += 20-A                         # adds black-and-black chips to the chips that land black face up

    return black        

  
  N = np.zeros([100, len(A_arr)])                 
  mean = np.zeros(len(A_arr))

  for j in range(len(A_arr)):             
    for i in range(100):                 # "shakes the box" 100 times
      N[i, j] = black_count(A_arr[j])    
    mean[j] = np.mean(N[:, j])
    x = N[:,j]
    plt.hist(x)
    plt.xticks(np.arange(21))
    plt.show()
  print("Mean: ", mean)
  #print(N)
```

```python
# Histogram plots for when the number of normal chips in a box of 20 chips is from 20 to 0
shake_box(np.arange(20, -1, -1))
```

```python
#Raw data: Normal setup (20 out of 20 chips are black-and-white chips)

row0 = np.array([11,6,11,14,8,8,11,5,11,9], int)
row1 = np.array([11,12,11,8,10,8,6,7,9,13], int)
row2 = np.array([11,9,14,8,12,13,6,6,9,13], int)
row3 = np.array([10,11,11,9,14,9,14,11,9,10], int)
row4 = np.array([14,8,13,11,16,12,10,7,9,7], int)
row5 = np.array([10,9,11,12,11,12,8,12,11,11], int)
row6 = np.array([8,10,12,11,10,7,10,9,10,10], int)
row7 = np.array([8,11,10,15,10,12,8,6,12,11], int)
row8 = np.array([10,11,9,13,9,8,13,9,11,10], int)
row9 = np.array([8,8,10,9,12,8,9,9,7,11], int)
data_normal = np.array([row0, row1, row2, row3, row4, row5, row6, row7, row8, row9], int)
x_normal = []
for i in range(10):
  for j in range(10):
    x_normal.append(data_normal[i,j])

#print(x_normal)
x_normal = np.array(sorted(x_normal), float)
print(x_normal)
```

```python
#Raw data: Biased setup (15 out of 20 chips are black-and-white chips; 5 out of 20 are black-and-black chips)

row0 = np.array([15,8,13,14,12,13,14,10,12,12], int)
row1 = np.array([13,13,15,13,12,12,9,11,12,14], int)
row2 = np.array([12,11,18,12,14,15,9,10,12,16], int)
row3 = np.array([12,13,12,13,16,12,16,12,12,12], int)
row4 = np.array([14,13,16,13,17,13,12,9,11,10], int)
row5 = np.array([13,12,12,15,15,14,9,13,14,14], int)
row6 = np.array([10,13,13,13,14,12,11,11,14,13], int)
row7 = np.array([13,14,13,16,14,14,11,11,14,14], int)
row8 = np.array([13,14,12,16,12,9,16,13,14,13], int)
row9 = np.array([11,12,12,12,15,10,11,12,12,14], int)
data_biased = np.array([row0, row1, row2, row3, row4, row5, row6, row7, row8, row9], int)
x_biased = []
for i in range(10):
  for j in range(10):
    x_biased.append(data_biased[i,j])

#print(x_biased)
x_biased = np.array(sorted(x_biased), float)
print(x_biased)
```

```python
# Data Treatment

Mean, Mode = 0,0
hist, bin_edges = 0,0
def data(x, Bins):
  Mean = statistics.mean(x)
  Mode = statistics.mode(x)
  STD = statistics.stdev(x)
  hist, bin_edges = np.histogram(x, Bins)
  frequencies = np.array(hist/100)

  return [Mean, Mode, STD, frequencies]

# Expected Statistical Distribution
def f_st(x):
  def F(n):
    return math.factorial(n)
  y = np.zeros(len(x))
  for i in range(21):
    y[i] = (F(20)/(F(20-x[i])*F(x[i])))*((0.5)**(20))
  return y
  
 
```

```python
data(x_normal, 12)
```

```python
plt.bar(np.linspace(5,16,12), data(x_normal, 12)[3],  color="blue", label="experimental distribution of unbiased chips", width=0.2)
plt.plot(np.linspace(0,20,21), f_st(np.linspace(0,20,21)), color="magenta", label="statistical distribution function, $f^{st}_j = 20!/[x_j!(20-x_j)!2^{20}]$")
plt.title("Normalized histogram of number of black faces in normal box")
plt.xticks(np.arange(21))
plt.legend(loc=[0, -0.4])
plt.xlabel('number of black faces')
plt.ylabel('distribution function')
plt.show()
```
